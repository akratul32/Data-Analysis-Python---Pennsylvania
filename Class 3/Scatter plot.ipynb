```python
#Visualization of review counts and star ratings of 'Health & Medical', 'Fast Food', 'Breakfast & Brunch'
import pandas as pd
xls=pd.ExcelFile('D:\Python practice\Data analysis using python\Mod 3 Code & Data Files\yelp.xlsx')
df=xls.parse('yelp_data')
allcities=xls.parse('cities')
allcities.head()
data_merge=pd.merge(left=df,right=allcities,how='inner',left_on='city_id',right_on='id')
data_merge.head()
allstates=xls.parse('states')
df_all=pd.merge(left=data_merge,right=allstates,how='inner',left_on='state_id',right_on='id')
del df_all['id_x']
del df_all['id_y']
df_all.head()
```

```python
df_health=df_all[df_all['category_0']=='Health & Medical']
df_fast=df_all[df_all['category_0']=='Fast Food']
df_break=df_all[df_all['category_0']=='Breakfast & Brunch']
df_break.head()
```

```python
%pylab inline
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.scatter(
    df_health['stars'], df_health['review_count'],
    marker='o',
    color='r',
    alpha=0.7,
    s=124,
    label='Health & Medical',
)

plt.scatter(
    df_fast['stars'], df_fast['review_count'],
    marker='o',
    color='b',
    alpha=0.7,
    s=124,
    label='Fast Food',
)

plt.scatter(
    df_break['stars'], df_break['review_count'],
    marker='o',
    color='b',
    alpha=0.7,
    s=124,
    label='Breakfast & Brunch',
)

plt.xlabel('Rating')
plt.ylabel('Review Count')

plt.legend(loc='best')

axes=plt.gca()
axes.set_yscale('log')

plt.show()
```
