```python
# How do review patterns differ across different cities? Which city has a higher number of 5 star reviews or 1 star reviews, Pittsburgh or Las Vegas?
import pandas as pd
xls=pd.ExcelFile('D:\Python practice\Data analysis using python\Mod 3 Code & Data Files\yelp.xlsx')
df=xls.parse('yelp_data')
allcities=xls.parse('cities')
allcities.head()
data_merge=pd.merge(left=df,right=allcities,how='inner',left_on='city_id',right_on='id')
data_merge.head()
allstates=xls.parse('states')
df_all=pd.merge(left=data_merge,right=allstates,how='inner',left_on='state_id',right_on='id')
del df_all['id_x']
del df_all['id_y']
df_all.head()
```

```python
%pylab inline
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
df_pitt=df_all[df_all['city']=='Pittsburgh']
df_pitt.head()
```

```python
df_vegas=df_all[df_all['city']=='Las Vegas']
df_vegas.head()
```

```python
pitt_stars=df_pitt['stars']
vegas_stars=df_vegas['stars']
vegas_stars.head()
```

```python
plt.hist(
    pitt_stars,
    alpha=0.3,
    color='blue',
    label='Pittsburgh',
    bins='auto'
)
plt.hist(
    vegas_stars,
    alpha=0.3,
    color='red',
    label='Las Vegas',
    bins='auto'
)

plt.xlabel('Rating')
plt.ylabel('Number of Rating Scores')

plt.legend(loc='best')

plt.title('Review distribution of Pittsburgh and Las Vegas ')

plt.show()
```

```python
#Histogram side by side
plt.hist(
    [pitt_stars,vegas_stars],
    alpha=0.7,
    color=['red','blue'],
    label=['Pittsburgh','Las Vegas'],
    bins='auto'
)

plt.xlabel('Rating')
plt.ylabel('Number of Rating Scores')

plt.legend(loc='best')

plt.title('Review distribution of Pittsburgh and Las Vegas')

plt.show()
```
